# Creating views from DATAFRAMES
    loc and iloc
    Note - loc takes only name assigned to row/column as an Argument.
    Note - iloc takes only index as an Argument
    loc[:,:]   accepted [range,range]
    loc[[],:]  accepted [[list],range]
    loc[:,[]]  accepted [range,[list]]
    loc[[],[]] accepted [[list],[list]] 

```python
import numpy as np
import pandas as pd
ndf= pd.DataFrame(np.random.rand(334,5), index= np.arange(334))
ndf.columns= list("ABCDE") # assigning names to columns
```

```python
ndf
```

```python
ndf.loc[[1,2],['B','C']] # Only the required view will be created
```

```python
ndf # notice that out data frame is not changed
```

```python
ndf.loc[[1,2],:] # Rows 1,2 and all columns generated.
```

```python
ndf.loc[:,['B','C']] # Columns B,C and all rows generated.
```

```python
ndf
```

```python
ndf.loc[(ndf['A']<0.3) & (ndf['A']>0.1)] #returns values for specified condition.
```

# Dropping Rows/Columns 

```python
ndf.drop([0] # dropping zeroth row. By default axis is 0
```

```python
# ndf.drop(['A']) # This wont work as axis needs to be defined. By default the axis is set at 0.
# Program starts finding the value of A in rows i.e 0th axis.
ndf.drop(['A'], axis = 1) # Defining axis is mandatory while dropping columns; as default axis is 0.
```

```python
ndf # Note that only view was created and actual coulumns were not dropped from the original dataframe.
```

```python
ndf.drop(['A'], axis = 1,inplace= True)
# This command [inplace=True] will not just create a view bu also drop the column from original dataframe.
```

```python
ndf # Column A dropped due to inplace command
```

```python
ndf.drop([3,4,5,6],inplace= True) #dropped rows permanently using inplace
```

```python
ndf #notice the rows have been dropped 
```

```python
ndf.reset_index() 
#As index was changed due to dropping off the rows. Index restored in sequence.
# It also adds an extra column named 'index' which we dont want.
```

```python
ndf.reset_index(drop= True) #removes that extra generated column of index
```

```python
ndf #Notice original dataframe not affected
```

```python
ndf.reset_index(drop= True, inplace = True) # INPLACE
```

```python
ndf # Dataframe changed 
```

```python
ndf['B'].isnull() # Checking for any null values in B column
```

```python
ndf['B']= None #Changing the entire B column to None.
# Important Note: Always set the values using loc/iloc.
# Setting the values this way may sometimes return temporary View without actually affecting the original dataframe. 
```

```python
ndf
```

```python
ndf['B'].isnull() #Returned True as B is none
```

# Dataframe- drop_duplicates, subset , keep - first,last,false commands

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
df.dropna() # drops all the rows which are NA 
```

```python
df.dropna(how = 'all') # Removes record only if all values are NA
```

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
df.dropna(how = 'all', axis = 1) # Notice that toy column was removed completely as each cell contained NA. 
```

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
df.drop_duplicates(subset= ['name']) # Drops the duplicates. Takes arguements as subset. Axis doesn't work here.
```

```python
df.drop_duplicates(subset= ['name'], keep= 'first') # Drops the duplicates. Takes arguements as subset.Keeps the first row.
```

```python
df.drop_duplicates(subset= ['name'], keep= 'last') # Drops the duplicates. Takes arguements as subset.Keeps the last row.
```

```python
df.drop_duplicates(subset= ['name'], keep= False) # Drops all duplicates from the specified subset.
```
